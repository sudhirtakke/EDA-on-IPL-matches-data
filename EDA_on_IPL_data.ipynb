###  Initial observations:

- Summary of data types in this dataset:
 - <em>Numeric</em>: <strong>Id</strong>, <strong>season</strong>, <strong>win_by_runs</strong>, <strong>win_by_wickets</strong> (integer)
  - <em>Boolean</em>: <strong>dl_applied</strong>
 - <em>Object</em>: Everything else
- <strong>dl_applied</strong> has value of '1' for 19 matches. It means rain interruption might have caused 19 times during all these seasons.
- <strong>id</strong> field has some extreme values, showing from 7949 to 7953, these looks not correct 
- <strong>season</strong> field shows values ranging from year 2008 to 2018, it means 10 years data and count for matches looks increasing. Possibly this means with every season, number of teams have increased and hence number of matches kept on increasing.
- <strong>umpire3</strong>  field has many missing values. Ideally this field value should not make any effect on deciding winner of the matches so we will remove this column.
- <strong>win_by_runs and win_by_wickets</strong>  Wherever one field has a value, other field has zero value. This is correct as team will win either by runs or wickets.
- <strong>winner</strong>  field three missing values, it means match didnt not happen and hence no result. We need to remove these rows.

```python
#Remove umpire3 column as it have many missing values
ipl_data.drop('umpire3', axis =1, inplace = True)
```

```python
#replace blank city values with most occuring city values
ipl_data.city = ipl_data.city.fillna(ipl_data['city'].mode()[0])
```

```python
#Check different team names and verify if they are all correct
print("Distinct teams are: ",len(set(ipl_data['team1'])))
print("Distinct team names are: ", set(ipl_data['team1']))
```

It can be seen that Pune team has three diffenet names - Pune warriors, Rising Pune Supergiant, Rising Pune supergiants.
we need to give only one name and replace other two values with this. 
Similarly for Sunrisers hyderabad team, there are three different names across different seasons. We will try to give only one name for Hyderabad team.

```python
#clean Pune team name

team1 = ipl_data['team1'].unique()
pune_team = ["Rising Pune Supergiants" ,"Pune Warriors", "Rising Pune Supergiant"]

for (row, col) in ipl_data.iterrows():
    
    if col.team1 in pune_team:
        ipl_data['team1'].replace(to_replace = col.team1, value = 'Rising Pune Supergiants', inplace = True)
    
    if col.team2 in pune_team:
        ipl_data['team2'].replace(to_replace = col.team2, value = "Rising Pune Supergiants", inplace = True)
        
    if col.winner in pune_team:
        ipl_data['winner'].replace(to_replace = col.winner, value = "Rising Pune Supergiants", inplace = True)
        
    if col.toss_winner in pune_team:
        ipl_data['toss_winner'].replace(to_replace = col.toss_winner, value = "Rising Pune Supergiants", inplace = True)

#clean Hyderabad team name

    if col.winner == 'Deccan Chargers' or col.winner == 'Hyderabad sunrisers':
        ipl_data['winner'].replace(to_replace = col.winner, value = "Sunrisers Hyderabad", inplace = True)
        
    if col.team1 == 'Deccan Chargers' or col.team1 == 'Hyderabad sunrisers':
        ipl_data['team1'].replace(to_replace = col.team1, value = "Sunrisers Hyderabad", inplace = True)
    
    if col.team2 == 'Deccan Chargers' or col.team2 == 'Hyderabad sunrisers':
        ipl_data['team2'].replace(to_replace = col.team2, value = "Sunrisers Hyderabad", inplace = True)
        
    if col.toss_winner == 'Deccan Chargers' or col.team2 == 'Hyderabad sunrisers':
        ipl_data['toss_winner'].replace(to_replace = col.toss_winner, value = "Sunrisers Hyderabad", inplace = True)      

print(ipl_data['team1'].unique())
print(ipl_data['team2'].unique())
print(ipl_data['winner'].unique())
print(ipl_data['toss_winner'].unique())
print(len(set(ipl_data['toss_winner'])))

```

#### Review whether missing data is still present

```python
#missing data
total = ipl_data.isnull().sum().sort_values(ascending=False)
percent = (ipl_data.isnull().sum()/ipl_data.isnull().count()).sort_values(ascending=False)
missing_data = pd.concat([total, percent], axis=1, keys=['Total', 'Percent'])
missing_data.head(20)
print(missing_data)
```

#### It can be seen that missing data is almost eliminated

```python
#Check different winners and verify if they are all correct
print("Distinct winning teams are: ",len(set(ipl_data['winner'])))
print("Distinct team names are: ", set(ipl_data['winner']))
```

Now we have sorted team names in columns team1, team2, winner based on cities and values are distinct ones
Although we have sorted with unique team1 and team2 values, winner column still has some nan values (3)

```python
#drop rows that are having winner as nan because these matches are not played at all
ipl_data.dropna(subset =['winner'])
```

### 4.3 Select sample of CSK team to study its data for winning matches

```python
# Select data only for Chennai Super kings matches from original data

ipl_data_csk1 = ipl_data[ipl_data.team1 == 'Chennai Super Kings']
ipl_data_csk2 = ipl_data[ipl_data.team2 == 'Chennai Super Kings']

ipl_data_csk = pd.concat([ipl_data_csk1, ipl_data_csk2], axis = 0) 
ipl_data_csk.count()
```

#### from this data, we will try to analyze different aspects of CSK team performance based on homeground advantage, toss winning, choosing bat or fields, opposition teams etc acroos all seasons


We need to add new columns for analysis as per above as existing data columns would not be enough

```python
ipl_data_csk['homeground']=ipl_data_csk.apply(lambda x: ' ' , axis=1)
ipl_data_csk['win']=ipl_data_csk.apply(lambda x: ' ' , axis=1)
ipl_data_csk['tosswin']=ipl_data_csk.apply(lambda x: ' ' , axis=1)
def func(x):
    if x == 'Chennai':
        return 'yes'
    else:
        return 'no'

def funct(x):
    if x == 'Chennai Super Kings':
        return 1
    else:
        return 0

def funct1(x):
    if x == 'Chennai Super Kings':
        return 'yes'
    else:
        return 'no'
    
ipl_data_csk['homeground'] = ipl_data_csk['city'].apply(func)
ipl_data_csk['win'] = ipl_data_csk['winner'].apply(funct)
ipl_data_csk['tosswin'] = ipl_data_csk['toss_winner'].apply(funct1)

ipl_data_csk
```

### 4.8 Against which opposition teams CSK team has better and worst performance in terms of match wins ?

Lets try to create a new column as opposition to include name of opposition team against CSK for all seasons

```python
ipl_data_csk['opposition'] = ipl_data_csk.apply(lambda x: x['team1'] if x['team2'] == 'Chennai Super Kings' else x['team2'],axis=1)
ipl_data_csk
```


### Winning performance of CSK team against all opposition teams

```python
relation = pd.crosstab(ipl_data_csk.opposition, ipl_data_csk.win )
relation.plot.barh(figsize=(10,5))
plt.xticks(size = 10)
plt.yticks(size = 10)
plt.title('Relation Between opposition team and wins for team CSK',size=25)
```

It can be observed that CSK team has better performance against Sunrisers Hyderabad team as total wins are 14 which is highest compared to other opposition teams and only 6 matches are lost.
Similarly CSK team has comparatively worst perfomance against Mumbai Indians as they have lost more matches than wins compared to other opposition teams.
